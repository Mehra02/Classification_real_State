```python
Data_Y=np.vstack(Data['grade'].values) # Selecting the target value as Data_Y whic is a datframe of target variable 'grade'
```

```python
print("Data_X Shape:",Data_X.shape)
print("Data_Y shape:",Data_Y.shape)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
Train_X,Test_X,Train_Y,Test_Y=train_test_split(Data_X,Data_Y,test_size=0.3,random_state=50) # Spliting the data into test and train
```

```python
print("Train_X Shape:",Train_X.shape,"Train_Y:",Train_Y.shape)
print("Test_X Shape:",Test_X
      
      .shape,"Test_Y:",Test_Y.shape)
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
Log_Model=LogisticRegressionCV(class_weight='balanced', multi_class='multinomial', solver='lbfgs') # Defining the logistic Regression with arguments that will be crucial for multiclass classification
```

```python
Log_Model.fit(Train_X,Train_Y) # Training the model  with the Train Data
```

```python
Train_Y_pred_Log=Log_Model.predict(Train_X) # After training the model has to predict the target values for training data
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score
```

```python
print("Accuracy_Score for the training data set :",accuracy_score(Train_Y,Train_Y_pred_Log))
# Checking the  accuracy Score for the Train Data that is observed 91.4% 
```

```python
Test_Y_pred_Log=Log_Model.predict(Test_X) # Predicting the values for Test Data 
```

```python
print("Accuracy_Score for for the Test Data set:",accuracy_score(Test_Y,Test_Y_pred_Log)) # Checking the accuracy score for the Test Data that is 84.07%
```
